## EP:
- 通过math库，写一个程序，使得用户输入三个顶点(x,y)返回三个角度
- 注意：Python计算角度为弧度制，需要将其转换为角度
<img src="../Photo/18.png">

```python
import math
x1,y1,x2,y2,x3,y3 = eval(input('输入三个顶点'))
a = math.sqrt((x2-x3)**2 + (y2-y3)**2)
b = math.sqrt((x3-x1)**2 + (y3-y1)**2)
c = math.sqrt((x2-x1)**2 + (y2-y1)**2)
A = math.acos((a*a - b*b - c*c) / (-2 *b*c))
B = math.acos((b*b - a*a - c*c) / (-2 *a*c))
C = math.acos((c*c - b*b - a*a) / (-2 *a*b))
print(math.degrees(A),math.degrees(B),math.degrees(C)) #将弧度转换成角度
```

## EP:
- 利用ord与chr进行简单邮箱加密

```python
title = "47845@qq.com"
result = ' '
for i in title:
    result = result + chr(ord(i)+10)
print(result)   #输出的结果在一行显示
                                    
```

```python
import hashlib
#md5  待加密的文本
str_ = "qwerty"
#创建MD5对象
h1 = hashlib.md5()
h1.update(str_.encode(encoding='utf-8'))  #可不写
print('MD5加密前为:'+str_)
print('MD5加密后为:'+h1.hexdigest())
```

```python
#输入密码，加密后与数据库中加密的密文密码对比 
import hashlib
dataset = "d8578edf8458ce06fbc5bb76a58c5ca4"
password = input('输入密码')
h1 = hashlib.md5()
h1.update(password.encode(encoding='utf-8'))
if dataset == h1.hexdigest():
    print('success')
else:
    print('false')
```
